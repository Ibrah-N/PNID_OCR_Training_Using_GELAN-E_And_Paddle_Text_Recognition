```python
# parse labels_2 file to rotate the vertical
# images into horizontal images because the
# paddle ocr models prefer horizontal images
# which gives high accuracy more generalization


import cv2
import numpy as np

with open("/content/dataset/labels_2.json") as file_parse:
  labels_parse = json.load(file_parse)


for idx, line_1 in enumerate(labels_parse):
  img = cv2.cvtColor(cv2.imread(f"/content/dataset/train/{line_1['image_path'].split('/')[-1]}"), cv2.COLOR_BGR2RGB)

  img_shape = (img.shape[0], img.shape[1]+15)
  if np.argmax(img_shape) == 0:
    img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
  else:
    continue

  # save the rotated images
  cv2.imwrite(f"/content/dataset/train/{line_1['image_path'].split('/')[-1]}", img)
```

```python
# it convert the label_1.json file int label.txt
# file which is supported by the paddleOCR recognition model
# label.txt format (img_path : "path/img.jpg", Transcription: "Hello")

import json


############ train #############
# convert json into .txt
file_type = "train" # "train"
with open("/content/dataset/labels_2.json", "r") as f:
  data = json.load(f)

for idx, line in enumerate(data):
  if '\n' in line['text']:
    continue
  else:
    with open(f"/content/dataset/{file_type}.txt", "a") as f:
      f.write(f"./{file_type}/{line['image_path'].split('/')[-1]}" + "\t" + line['text'] + "\n")



############ test #############
# convert json into .txt
file_type = "val" # "test"
with open("/content/dataset/test_1.json", "r") as f:
  data = json.load(f)

for idx, line in enumerate(data):
  if '\n' in line['text']:
    continue
  else:
    with open(f"/content/dataset/{file_type}.txt", "a") as f:
      f.write(f"./{file_type}/{line['image_path'].split('/')[-1]}" + "\t" + line['text'] + "\n")
```
